### Let's Check correlation between toss winner and match winner

```python
from scipy.stats import chi2_contingency
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['toss_winner'], df['winner'])

# Calculate Chi-Square statistic, p-value, degrees of freedom, and expected frequencies
chi2, p, dof, ex = chi2_contingency(contingency_table)

# Calculate CramÃ©r's V
n = contingency_table.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))

print(f"CramÃ©r's V: {cramers_v}")

# Visualize the contingency table
sns.heatmap(contingency_table, annot=True, fmt="d", cmap="Blues")
plt.title("Contingency Table of Toss Winner vs Winner")
plt.xlabel("Matches Won")
plt.ylabel("Toss Winner")
plt.show()
```

## Analysis of the Table
* ### Diagonal Elements: The cells along the diagonal (top-left to bottom-right) show the number of times the team that won the toss also won the match. For instance, Bangalore (row 1, column 1) won 4 matches after winning the toss.

* ### Off-Diagonal Elements: These cells represent the matches where a team won the toss but a different team won the match. For instance, Bangalore (row 1) winning the toss and Punjab (column 9) winning the match has a count of 1

### **Key Observations**
 ### Bangalore:

* ### Won the toss and match 4 times (diagonal cell value).
* ### Won the toss but lost the match to different teams multiple times.

### Hyderabad:

* ### Won the toss and match 3 times.

* ### Notably, Hyderabad has higher values in matches won by Kolkata and Lucknow when Hyderabad won the toss, indicating possible difficulties in converting toss wins to match wins against these teams.

### Rajasthan:

* ### Despite winning the toss multiple times, Rajasthan has a significantly high number of match wins (7) when they won the toss, showing a strong conversion rate from toss wins to match wins.

###General Trends:

* ### Teams like Kolkata and Punjab have some consistency in converting toss wins to match wins but less frequently than Rajasthan or Hyderabad.

* ### There are instances of toss winners failing to convert the advantage into a match win, seen in off-diagonal cells with non-zero values.

### Color Coding
* ### Color Intensity: The color intensity of each cell corresponds to the number of matches. Darker shades represent a higher count, while lighter shades represent a lower count. For example, Rajasthanâ€™s dark blue cell (7) indicates a high number of match wins when they won the toss.

### Interpretation
* ### Winning the Toss Advantage: The graph can be used to assess whether winning the toss provides a significant advantage in terms of winning the match. Teams with high diagonal values relative to off-diagonal values (like Rajasthan) seem to benefit more from winning the toss.

* ### Strategic Insights: Teams can use this data to strategize better. For example, teams with low diagonal values might need to focus on converting toss wins into match wins.
